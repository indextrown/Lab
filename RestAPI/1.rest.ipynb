## 추천 리스트(No DTO)

```python
import requests
import json
from dotenv import load_dotenv
import os

# ✅ .env 로드
load_dotenv()
BASE_URL = os.getenv("API_BASE_URL")

# ✅ URL 동적으로 구성
url = f"{BASE_URL}/api/v1/recommend"

# ✅ GET 요청
response = requests.get(url)

# ✅ 상태 코드 출력
print("Status Code:", response.status_code)

# ✅ JSON 보기 좋게 출력
try:
    data = response.json()
    pretty = json.dumps(data, indent=4, ensure_ascii=False)
    print("Response JSON:", pretty)
except Exception as e:
    print("❌ Response Text:", response.text)
```

## 추천 리스트

```python
import requests
import json
from dataclasses import dataclass
from typing import List

# ✅ DTO 정의
@dataclass
class RecommendDTO:
    id: int
    recommendName: str

# ✅ API 요청
url = f"{BASE_URL}/api/v1/recommend"
response = requests.get(url)
print("Status Code:", response.status_code)

try:
    data = response.json()
    
    # ✅ 응답 데이터를 DTO 리스트로 변환
    recommend_list: List[RecommendDTO] = [RecommendDTO(**item) for item in data]

    # ✅ 이쁘게 출력
    print("📦 전체 추천 카테고리 목록:")
    for r in recommend_list:
        print(f"👉 ID: {r.id} | 이름: {r.recommendName}")

except Exception as e:
    print("❌ Response Text:", response.text)
```

## PopupDTO

```python
from dataclasses import dataclass
from typing import Optional

@dataclass
class PopupDTO:
    id: int
    popupUuid: str
    name: str
    startDate: str
    endDate: str
    openTime: Optional[str]       # ⬅ 문자열로 변경
    closeTime: Optional[str]      # ⬅ 문자열로 변경
    address: str
    roadAddress: str
    region: str
    latitude: float
    longitude: float
    geocodingQuery: str
    instaPostId: str
    instaPostUrl: str
    likeCount: int
    captionSummary: Optional[str]
    caption: Optional[str]
    imageUrl: Optional[str]
    mediaType: Optional[str]
    errorCode: Optional[str]

    # ✅ 헬퍼 메서드
    def display(self):
        print(f"📌 {self.name} ({self.region})")
        print(f"🗓️ {self.startDate} ~ {self.endDate}")
        print(f"📍 {self.roadAddress}")
        if self.openTime and self.closeTime:
            print(f"🕒 {self.openTime} ~ {self.closeTime}")
        print(f"❤️ 좋아요 {self.likeCount}")
        print(f"📸 이미지 URL: {self.imageUrl}")
        print(f"📖 본문: {self.captionSummary}")
        print("-" * 40)
```

## 팝업 전체조회

```python
import requests
from dataclasses import dataclass
from typing import List, Optional

# ✅ 위 DTO 정의 복붙 (TimeDTO, PopupDTO)
url = f"{BASE_URL}/api/v1/popup"

response = requests.get(url)
print("Status Code:", response.status_code)

try:
    data = response.json()
    popup_list: List[PopupDTO] = []

    for item in data:

        popup = PopupDTO(
            id=item["id"],
            popupUuid=item["popupUuid"],
            name=item["name"],
            startDate=item["startDate"],
            endDate=item["endDate"],
            openTime=item.get("openTime"),   # 문자열
            closeTime=item.get("closeTime"), # 문자열
            address=item["address"],
            roadAddress=item["roadAddress"],
            region=item["region"],
            latitude=item["latitude"],
            longitude=item["longitude"],
            geocodingQuery=item["geocodingQuery"],
            instaPostId=item["instaPostId"],
            instaPostUrl=item["instaPostUrl"],
            likeCount=item["likeCount"],
            captionSummary=item.get("captionSummary"),
            caption=item.get("caption"),
            imageUrl=item.get("imageUrl"),
            mediaType=item.get("mediaType"),
            errorCode=item.get("errorCode")
        )
        popup_list.append(popup)

    # ✅ 결과 출력
    print(f"총 {len(popup_list)}개의 팝업이 조회되었습니다.\n")
    for p in popup_list:
        p.display()

except Exception as e:
    print("❌ Response Text:", response.text)
```

## 알림 키워드 조회

```python
from dataclasses import dataclass
from typing import List
import requests

# ✅ DTO 정의
@dataclass
class AlertKeywordDTO:
    alertKeyword: str

# ✅ GET 요청 함수
def get_alert_keywords(user_uuid: str) -> List[AlertKeywordDTO]:
    url = f"{BASE_URL}/api/v1/alert-keyword"
    params = {"userUuid": user_uuid}
    response = requests.get(url, params=params)
    print("Status Code:", response.status_code)

    data = response.json()
    keywords = [AlertKeywordDTO(**item) for item in data]

    print(f"총 {len(keywords)}개의 키워드가 조회되었습니다.")
    for k in keywords:
        print(f"🔔 {k.alertKeyword}")
    return keywords

# ✅ 예시 호출
get_alert_keywords("e2ca7e3d-dff8-4a22-ac11-6ad4a9c2d274")
```

## 알림 키워드 등록

```python
from dataclasses import dataclass, asdict
import requests

# ✅ POST용 DTO
@dataclass
class CreateAlertKeywordDTO:
    userUuid: str
    newAlertKeyword: str

def create_alert_keyword(user_uuid: str, keyword: str):
    url = f"{BASE_URL}/api/v1/alert-keyword"
    dto = CreateAlertKeywordDTO(userUuid=user_uuid, newAlertKeyword=keyword)
    payload = asdict(dto)
    headers = {"Content-Type": "application/json"}

    response = requests.post(url, json=payload, headers=headers)
    print("Status Code:", response.status_code)
    print("Response:", response.text)

# ✅ 예시 호출
create_alert_keyword("e2ca7e3d-dff8-4a22-ac11-6ad4a9c2d274", "스타벅스")
```

## 알림 키워드 삭제

```python
from dataclasses import dataclass, asdict
import requests

# ✅ DELETE용 DTO
@dataclass
class DeleteAlertKeywordDTO:
    userUuid: str
    deleteAlertKeyword: str

def delete_alert_keyword(user_uuid: str, keyword: str):
    url = f"{BASE_URL}/api/v1/alert-keyword"
    dto = DeleteAlertKeywordDTO(userUuid=user_uuid, deleteAlertKeyword=keyword)
    payload = asdict(dto)
    headers = {"Content-Type": "application/json"}

    response = requests.delete(url, json=payload, headers=headers)
    print("Status Code:", response.status_code)
    print("Response:", response.text)

# ✅ 예시 호출
delete_alert_keyword("e2ca7e3d-dff8-4a22-ac11-6ad4a9c2d274", "스타벅스")
```
